Входните данни са разпределени в три категории: "clean", "dirty" и "borderline". Те представляват съответно щайги в различните им състояния - чисти, мръсни и гранични(трудни за класифициране). Разпределени са по следния начин (в брой снимки):

```python
import os
import matplotlib.pyplot

pictures_count_list = [len(os.listdir("../data/clean")), len(os.listdir("../data/dirty")), len(os.listdir("../data/borderline"))]

matplotlib.pyplot.bar(["clean", "dirty", "borderline"], pictures_count_list)

matplotlib.pyplot.ylabel("Pictures count")

matplotlib.pyplot.show()
```

Примери за трите категории(в реда - "clean", "dirty", "borderline"):

```python
import os
import matplotlib.pyplot
import matplotlib.image

files_paths = ["../data/clean/" + os.listdir("../data/clean")[0],
               "../data/dirty/" + os.listdir("../data/dirty")[0],
               "../data/borderline/" + os.listdir("../data/borderline")[0]]

files_paths_count = len(files_paths)

labels = ["clean", "dirty", "borderline"] # len should be the same as files_paths_count

for i in range(0, files_paths_count):
    
    image = matplotlib.image.imread(files_paths[i])
    
    matplotlib.pyplot.title(labels[i], fontsize = 20)
    
    matplotlib.pyplot.imshow(image)
    
    matplotlib.pyplot.axis("off")
    
    matplotlib.pyplot.show()
```

На база горепосочените примери и броят на данните, може да се установи, че правилна класификация може да се направи, като се приложи машинно самообучение(най - добре спецеиализиран вариант - примерно CNN(convolutional neural network)). Но тъй като разработката на това решение изисква голямо количество ресурси и време, ще направя опит за класификация, който не прилага такива модели, след което ще сравня двете решения и ще изведа положителните и отрицателните четри на всяко едно.
